```python
image = [['N','N','N','B','B'],
         ['B','B','N','N','N'],
         ['N','N','B','B','B'],
         ['B','N','N','N','B'],
         ['B','B','N','B','B']]
```

```python
image_compress = ""
```

```python
def compress_line(line_image):
    j = 0
    line_compress = ""
    while j<5:
        compt = 0
        pix = line_image[j]
        while j<5 and pix == line_image[j]:
            j = j+1
            compt = compt+1
        if compt == 1:
            nb = ""
        else :
            nb = str(compt)
        line_compress = line_compress+nb+pix
    line_compress = line_compress + '0'
    return line_compress
```

```python
compress_line('NNNNN')
```

```python
compress_line('NNNBB')
```

```python
compress_line('BBNBB')
```

```python
for i in range(len(image)):
    line = image[i]
    image_compress = image_compress + compress_line(line)
print(image_compress)
```

```python
def compress_line(line_image):
    j = 0
    line_compress = ""
    while j<len(line_image):
        compt = 0
        pix = line_image[j]
        while j<len(line_image) and pix == line_image[j]:
            j = j+1
            compt = compt+1
        if compt == 1:
            nb = ""
        else :
            nb = str(compt)
        line_compress = line_compress+nb+pix
    line_compress = line_compress + '0'
    return line_compress

image = [  15*['N'] + 25 * ['B'], 
           5*['N'] + 5 * ['B']+ 5*['N'],
         2 * ['B']+ 50*['N']]

image_compress = ""
for i in range(len(image)):
    line = image[i]
    image_compress = image_compress + compress_line(line)
print(image_compress)
```
