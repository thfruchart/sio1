# Exercice 2

```python
n = 5
while n > 1 :
    r = n%2
    if r == 0:
        n = n // 2
    else : 
        n = 3*n+1
    print(n)
print('terminé')
```

# Exercice 5

```python
STOP = 100
meilleur_n = 0
plus_grand_nb_affiche = 0 
essai = 1
while essai <= STOP :
    n = essai
    cpt = 0
    while n > 1 :
        r = n%2
        if r == 0:
            n = n // 2
        else : 
            n = 3*n+1
        #print(n)
        cpt = cpt + 1
    #print('nombre de valeurs : ', cpt)
    #print('n=',essai,':', cpt, 'affichages')
    if cpt >= plus_grand_nb_affiche : 
        print('n=',essai,':', cpt, 'affichages')
        plus_grand_nb_affiche = cpt
        meilleur_n = essai
    essai = essai + 1
    #print('-----')
print()
print('meilleur n : ', meilleur_n)
```
