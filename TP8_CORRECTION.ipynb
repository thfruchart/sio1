### Principe du chiffrement affine

        LETTRE →    nombre  –  –   –  →   nombre    → LETTRE 
         'A'   →     0      –  –   –  →    17       →   'R'
         'Z'   →     25     –  –   –  →    12   	→   'M'

         car   →     x      –  –   –  →    r        →   code_lettre(car)

```python
ALPHABET = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
a = 5
b = 17

def RANG(car):
    resu = -1
    i = 0
    while (i<26) and resu == -1:
        if (car ==ALPHABET[i]):
            resu = i
        i = i+1
    return resu
```

```python
def decode_lettre1(car):
    resu = car
    for i in range(26):
        origine=ALPHABET[i]
        if code_lettre(origine)==car:
            resu = origine
    return resu


def decode_message1(txt):
    resu=''
    longueur = len(txt)
    for i in range(longueur):
        lettre = txt[i]
        resu=resu + decode_lettre1(lettre)
    return resu
```
