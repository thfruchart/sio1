## Somme de deux matrices


##### question 2 
Compléter le code suivant pour qu'il puisse calculer la somme de deux matrices de même format.

```python
def somme(M,N):
    nblignesM = len(M)
    nbcolM = len(M[0])
    nblignesN = len(N)
    nbcolN = len(N[0])
    
    #if  not( nblignesM == nblignesN  and  nbcolM == nbcolN  ):
    if ( nblignesM != nblignesN  or  nbcolM != nbcolN  ):
        print("Matrices de dimensions differentes, calcul impossible !")
    else : 
        resu =[[0 for j in range(nbcolM)] for i in range(nblignesM)]
        for i in range(nblignesM):
            for j in range(nbcolM):
                resu[i][j]= M[i][j] + N[i][j]  # A COMPLETER
        return resu        
```

Tester la fonction sur les matrices définies précédemment

```python
somme(A1,A2)
```

```python
somme(C,C)
```

```python
somme(L,C)
```

## Produit de deux matrices

##### question 3
Compléter le code suivant pour qu'il puisse calculer le produit de deux matrices de formats compatibles.

```python
def produit_matrices(M,N):
# tous les ... sont à compléter
    n=len(M)    # nombre de lignes de M
    p=len(M[0]) # nombre de colonnes de M
    q=len(N[0]) # nombre de colonnes de N
    if (len(N) == p):
        resu=[ [ 0 for j in range(q)] for i in range(n)]
        for i in range(n):
            for j in range(q):
                for k in range(p) : 
                    resu[i][j] += M[i][k] * N[k][j]
        return resu
    else : 
        print("Erreur de format")
```

Tester la fonction sur les matrices définies précédemment

```python
produit_matrices(A1,C)
```

```python
produit_matrices(A1,A2)
```
